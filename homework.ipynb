```python
students = []
n = int(input("Enter number of students:"))

subjects = ["English","Math","Nepali","Science"]

for i in range(n):
    marks = []
    name = input("Enter name: ")
    for sub in subjects:
        mark = int(input(f"Enter marks in {sub}: "))
        marks.append(mark)

    total = sum(marks)
    percentage = total / len(subjects)

    if any(m < 32 for m in marks):
        status = "Fail"
        grade  = "No Grade"
    else:
        status = "PASS"

        if percentage >= 80:
            grade = "A+"
        elif percentage >= 70:
            grade = "A"
        elif percentage >= 60:
            grade = "B"
        elif percentage >= 50:
            grade = "C"
        elif percentage >= 40:
            grade = "D"
        else:
            grade = "F"

    students.append({
                "name":name,
                "English":marks[0],
                "Math":marks[1],
                "Nepali":marks[2],
                "Science":marks[3],
                "total":total,
                "percentage":percentage,
                "status":status,
                "grade":grade
            })
    
print("\n\n============= Student Result Sheet =================\n")  
for s in students:
    print("Name       :",s["name"])
    print("English    :",s["English"])
    print("Maths      :",s["Math"])
    print("Nepali     :",s["Nepali"])
    print("Science    :",s["Science"])
    print("Total      :",s["total"])
    print("Percentage :",s["percentage"])
    print("Status     :",s["status"])
    if s["status"]=="PASS":
         print("Grade      :",s["grade"])
    else:
         print("Grade      :  No grade(failed)")
    print("-"*40)
        
```
