```python
students={
    "ram":{"maths":80,"science":80,"english":80},
    "shyam":{"maths":80,"science":80,"english":80},
    "hari":{"maths":80,"science":80,"english":80}
}

for name,subjects in students.items():
    total= sum(subjects.values())
    avg = total/len(subjects)

    if avg>=80:
        grade="A+"
    elif avg>=60:
        grade="b"
    else:
        grade="c"

    print(f"{name}: Average: {avg:.2f}, Grade:{grade}")
```
