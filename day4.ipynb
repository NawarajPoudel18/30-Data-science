```python
students={
    "ram":{"maths":80,"science":80,"english":80},
    "shyam":{"maths":80,"science":80,"english":80},
    "hari":{"maths":80,"science":80,"english":80}
}

for name,subjects in students.items():
    total= sum(subjects.values())
    avg = total/len(subjects)

    if avg>=80:
        grade="A+"
    elif avg>=60:
        grade="b"
    else:
        grade="c"

    print(f"{name}: Average: {avg:.2f}, Grade:{grade}")
```

```python
import time

a = 3

users = {
    "admin": "1234",
    "nawaraj": "12345678",
    "hari": "987654321"
}

while a > 0:
    username = input("Enter your username: ")
    password = input("Enter your password: ")

    if username in users and users[username] == password:
        print("\033[92mLogin Successful\033[0m")
        break
    else:
        a -= 1
        print(f"\033[91mInvalid username/password. {a} attempts remaining\033[0m")

        if a > 0:
            print("Wait 5 seconds before retry...")
            time.sleep(1)

if a == 0:
    print("\033[91mAccount locked. No attempts left.\033[0m")
```

```python
cart = {
    "apple":{"price":300, "qty":2},
    "bannana":{"price":360, "qty":2},
    "milk":{"price":70, "qty":2}
}
total = 0

for item, details in cart.items():
    cost = details["price"]*details["qty"]
    total+=cost
    print(item,":",cost)

print("total bill:",total)
```
