```python

import numpy as np

print("NumPy Assignment Solutions")
```

## Section A: Data Types

```python

# Q1
arr_int = np.array([1, 2, 3], dtype="int32")
arr_float = np.array([1.1, 2.2, 3.3], dtype="float64")
arr_bool = np.array([True, False, True], dtype="bool")
arr_date = np.array(['2026-01-01'], dtype='datetime64')

print(arr_int)
print(arr_float)
print(arr_bool)
print(arr_date)

# Q2
arr = np.array([1, 2, 3, 4], dtype=np.float64)
print(arr)

# Q3
arr_dtype = np.array([1.5, 2.5, 3.5])
print(arr_dtype.dtype)

# Q4
dates = np.arange('2026-01-01', '2026-01-08', dtype='datetime64[D]')
print(dates)

# Q5
print("dtype is important because it controls memory usage, speed, and calculations.")
```

## Section B: Vectorization

```python

# Q1
prices = np.array([100, 200, 300, 400])
taxed_prices = prices * 1.13
print(taxed_prices)

# Q2
salaries = np.array([30000, 50000, 70000])
new_salaries = salaries * 1.10
print(new_salaries)

# Q3
celsius = np.array([0, 20, 30, 40])
fahrenheit = (celsius * 9/5) + 32
print(fahrenheit)

# Q4
discounted = prices * 0.80
print(discounted)

# Q5
salary = np.array([40000, 55000, 70000])
bonus = np.where(salary > 50000, salary * 0.15, salary * 0.05)
print(bonus)
```

## Section C: Broadcasting

```python

matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

# Q1
print(matrix + 10)

# Q2
vector = np.array([10, 20, 30])
print(matrix + vector)

# Q3
column_vector = np.array([[2], [3]])
print(matrix * column_vector)

# Q4
row_vector = np.array([1, 2, 3])
print(matrix + row_vector)

# Q5
print("Broadcasting allows NumPy arrays with different shapes to work together.")
```

## Section D: Mathematical Functions

```python

sales = np.array([120, 140, 200, 250, 300])

# Q1
print("Sum:", np.sum(sales))
print("Mean:", np.mean(sales))
print("Median:", np.median(sales))

# Q2
print("Standard Deviation:", np.std(sales))
print("Variance:", np.var(sales))

# Q3
print("Min:", np.min(sales))
print("Max:", np.max(sales))

# Q4
print("Square Root:", np.sqrt(sales))

# Q5
print("Log Values:", np.log(sales))

# Q6
print("Cumulative Sum:", np.cumsum(sales))
```

## Section E: Axis

```python

arr = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

# Q1
print("Row-wise Sum:", np.sum(arr, axis=1))

# Q2
print("Column-wise Sum:", np.sum(arr, axis=0))

# Q3
print("Row-wise Mean:", np.mean(arr, axis=1))

# Q4
print("Column-wise Max:", np.max(arr, axis=0))

# Q5
print("axis=0 works column-wise")
print("axis=1 works row-wise")
```

## Section F: Filtering & Conditions

```python

marks = np.array([45, 76, 88, 34, 92, 60])

# Q1
print(marks[marks > 60])

# Q2
print(marks[marks < 50])

# Q3
updated_marks = np.where(marks < 40, 40, marks)
print(updated_marks)

# Q4
result = np.where(marks >= 40, "Pass", "Fail")
print(result)

# Q5
print(marks[marks % 2 == 0])
```

## Section G: Percentile, Quantile, Correlation

```python

data = np.array([10, 20, 30, 40, 50, 60, 70])

# Q1
print("25th Percentile:", np.percentile(data, 25))

# Q2
print("Median:", np.percentile(data, 50))

# Q3
print("Quantile 0.75:", np.quantile(data, 0.75))

# Q4
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

correlation = np.corrcoef(x, y)
print(correlation)

# Q5
print("Correlation close to 1 means strong positive relationship.")
```

## Section H: Real-world Case Study

```python

sales = np.array([
    [200, 300, 250],
    [400, 500, 450],
    [150, 250, 350]
])

# Q1
print("Total Sales Per Store:", np.sum(sales, axis=1))

# Q2
print("Total Sales Per Month:", np.sum(sales, axis=0))

# Q3
print("Average Monthly Sales:", np.mean(sales, axis=0))

# Q4
print("Highest Sales:", np.max(sales))

# Q5
store_totals = np.sum(sales, axis=1)
print("Stores with sales > 1000:", store_totals[store_totals > 1000])
```

## Mini Project: Restaurant Sales Analysis

```python

sales = np.array([
    [1200, 1500, 1100, 1700],
    [1000, 1300, 1400, 1600],
    [900, 1100, 1000, 1500]
])

# Task 1
branch_sales = np.sum(sales, axis=1)
print("Total Sales Per Branch:", branch_sales)

# Task 2
branches = ["Branch A", "Branch B", "Branch C"]
highest_branch = branches[np.argmax(branch_sales)]
print("Highest Branch:", highest_branch)

# Task 3
print("Average Weekly Sales:", np.mean(sales, axis=0))

# Task 4
weeks = ["Week1", "Week2", "Week3", "Week4"]
highest_week = weeks[np.argmax(np.sum(sales, axis=0))]
print("Week with Highest Revenue:", highest_week)

# Task 5
taxed_sales = sales * 1.13
print("Sales After 13% Tax:")
print(taxed_sales)
```
