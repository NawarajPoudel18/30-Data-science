```python
def c_to_f(c):
    return c*1.8+32
def f_to_c(f):
    return (f-32)/1.8
print(c_to_f(32))
print(f_to_c(89.6))
```

```python
def sum(n):
    total=0
    for i in n:
        total+=i
    return total
num=[1,2,3,4,5,6,7]
sum(num)
```

```python
def my_sum(*name):
    return sum(name)

print(my_sum(1,2,3,4,5,6,7,8,9))
```

```python
def my_sum(*clean):
    value=[]
    for s in clean:
        if s!=None:
             value.append(s)
    return value
my_sum(1,None,2,3,None,4,5,None,222)
```
