Calculator

```python
a = float(input("Enter your First Number"))
b = float(input("Enter your second Number"))
op = str(input("Enter a operator(+,-,*)"))
if op == "+":
    result = a + b
elif op == "-":
    result = a - b
elif op == "*":
    result = a * b

print(result) 
```
