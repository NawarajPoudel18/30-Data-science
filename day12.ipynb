```python
are=np.array([1,2,3,5,6,7,8,9])
add=are+3
sub=are-3
mul=are*2
div=are/2
print(f"the value after addition is :{add}")
print(f"the value after subtraction is :{sub}")
print(f"the value after multiplication is :{mul}")
print(f"the value after division is :{mul}")
sum=np.mean(are)
print(f"the mean is : {sum}")
print(f"The value in arrey greater then its mean are :{are[are>sum]}")
```
