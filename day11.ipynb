```python
import numpy as np

# i → integer
a1 = np.array([1, 2, 3, 4, 5], dtype="i")

# b → boolean
b1 = np.array([True, False, True], dtype="b")

# u → unsigned integer
# u1 = np.array([1, 2, 3, 4, 5], dtype="u")

# f → float
f1 = np.array([1.5, 2.7, 3.9], dtype="f")

# c → complex
c1 = np.array([1+2j, 2-3j], dtype="c")

# m → timedelta
m1 = np.array([np.timedelta64(5, 'D'), np.timedelta64(2, 'D')], dtype="m")

# M → datetime
M1 = np.array(["2025-12-21", "2026-01-01"], dtype="M")

# O → object
O1 = np.array([1, "hello", 3.5], dtype="O")

# S → byte string
S1 = np.array([b"hello", b"world"], dtype="S")

# U → unicode string
U1 = np.array(["Hello", "Nepal"], dtype="U")

# V → void (raw memory)
# V1 = np.array([(1, 2), (3, 4)], dtype="V")


# OUTPUT
print("i (int)        :", a1, "dtype:", a1.dtype)
print("b (bool)       :", b1, "dtype:", b1.dtype)
# print("u (unsigned)   :", u1, "dtype:", u1.dtype)
print("f (float)      :", f1, "dtype:", f1.dtype)
print("c (complex)    :", c1, "dtype:", c1.dtype)
print("m (timedelta)  :", m1, "dtype:", m1.dtype)
print("M (datetime)   :", M1, "dtype:", M1.dtype)
print("O (object)     :", O1, "dtype:", O1.dtype)
print("S (byte string):", S1, "dtype:", S1.dtype)
print("U (unicode)    :", U1, "dtype:", U1.dtype)
# print("V (void)       :", V1, "dtype:", V1.dtype)
```
