```python
n=6
for i in range(n):
    for j in range(n-i-1):
        print(" ", end=" ")
    for k in range(2*i+1):
        print("*",end=" ")
    print()
```

```python
n=6
for i in range(n):
    for j in range(n-i-1):
        print(" ", end=" ")
    for k in range(2*i+1):
        print("*", end=" ")
    print()
for i in range(n-2, -1, -1):
    for j in range(n-i-1):
        print(" ", end=" ")
    for k in range(2*i+1):
        print("*", end=" ")
    print()
```

```python
n=6
for i in range(n):
    for j in range(n-i-1):
        print(" ", end=" ")
    for k in range(2*i+1):
        print(j, end=" ")
    print()
for i in range(n-2, -1, -1):
    for j in range(n-i-1):
        print(" ", end=" ")
    for k in range(2*i+1):
        print(j, end=" ")
    print()
```

```python
n=6
for i in range(n):
    for j in range(n):
        if i==0 or  i==n-1 or j==0 or j==n-1 :
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# value on n should be odd 
n=5
for i in range(n):
    for j in range(n):
        if i==j or i+j==n-1 :
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```
