Patterns


```python
for i in range(10):
  for j in range (10):
     print("*",end="")
  print()
```

```python
for i in range(10):
    for j in range(i+1):
        print("*", end="")
    print()
```

```python
for i in range(1,6):
    for j in range(1,i+1):
        print(j,end="")
    print()
```

```python
for i in range(10,1,-1):
    for j in range(i):
        print("*", end="")
    print()
for i in range(10):
    for j in range(i+1):
        print("*", end="")
    print()
```

```python
n=10
for i in range(n,0,-1):
    print("*"*i)
for i in range(2,n+1):
    print("*"*i)
```

```python
n=6
for i in range(n):
    for j in range(n-i-1):
        print(" ", end=" ")
    for k in range(2*i+1):
        print("*",end=" ")
    print()
```

```python
n=6
for i in range(n):
    for j in range(n-i-1):
        print(" ", end=" ")
    for k in range(2*i+1):
        print("*", end=" ")
    print()
for i in range(n-2, -1, -1):
    for j in range(n-i-1):
        print(" ", end=" ")
    for k in range(2*i+1):
        print("*", end=" ")
    print()
```
