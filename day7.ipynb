lambda function


```python
square= lambda x :x*x
square(10)
```

```python
sum= lambda x,y:x+y
sum(1,2)
```

```python
sum_all= lambda *x :sum(x)
sum_all(10,11,34,456)
```

```python
check= lambda a: (a%2) ==0 #check if the value is even
print(check(2))
print(check(3))
```

```python
#check maximum
check_max= lambda a,b: a if a>b else b
#check minimum
check_min= lambda a,b: a if a<b else b
print(check_max(10,20))
print(check_min(10,20))
```

```python
check_max= lambda a,b,c: a if a>b and a>c else max(b,c)
check_max(4,3,3)

```
