```python
num=[1,2,3,4]
result=list(map(lambda x: x*x, num)) #list map make every veriable a function and help us to get the job done
print(result)
```

```python
sub=["nawaraj","sachendra","don"]
result=list(map(lambda x: x.upper(),sub))
print(result)
```

```python
numbers=[1,2,3,456,78,9,9]
result =list(map(lambda x: x+10,numbers))
print(result)
```

```python
number=[0,1,2,3,5,6,8,9,36,35,38,39,51]
filter=list(filter(lambda x: x>10,number))
print(filter)
```

```python
word=["python", "" ,"nawaraj", ""]
result=list(filter(lambda x: x!="",word))
print(result)
```

```python
letter = "n","N"

name = ["nawaraj","surendra","Nirajan","prabin","suraj"]

result = list(filter(lambda x: x.startswith(letter), name))

print(result)
```

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)
print(factorial(9))
```

```python
def sum_list(lst):
    if len(lst) == 0:
        return 0
    return lst[0] + sum_list(lst[1:])
print(sum_list([1,2,3,4,5]))
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(6))
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

num = int(input("Enter terms: "))

for i in range(num):
    print(fibonacci(i), end=" ")
```
